```python
import socket
import threading
import requests
```

```python
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
pars = ('127.0.0.1', 80)
s.bind(pars)
```

```python
s.listen(5)
```

```python
def serveClient(clientsocket, address):
    
    data = clientsocket.recv(1024)
    print("from client", data)

    data_decode = data.decode()
    print(data_decode[5:12])

    if data_decode[5:13] == 'get.html':
        responseData = b'<html><body>good.html</body></html>'
        clength = len(responseData)
        responseStatus = b'HTTP/1.0 200 OK\r\n'
        responseHeader = '%s%s%s%s%s%d%s' % ('Date : Mon, 07 Dec 2020 12:12:12 GMT\r\n',
                                            'Server : localhost\r\n',
                                            'connection : close\r\n',
                                            'Content-Type : text/html; charset=UTF-8\r\n',
                                            'Content-Length: ', clength, '\r\n')
        responseHeader_encode = responseHeader.encode(encoding = 'utf-8')

        clientsocket.send(responseStatus)
        clientsocket.send(responseHeader_encode)
        clientsocket.send(b'\r\n')
        clientsocket.send(responseData)


    if data_decode[5:18] == 'redirect.html':
        
        responseData = b''
        clength = len(responseData)
        responseStatus = b'HTTP/1.0 301 Moved Permanently\r\n'
     
        responseHeader = '%s%s%s%s%s%s%d%s' % ('Date : Mon, 07 Dec 2020 12:12:12 GMT\r\n',
                                            'Server : localhost\r\n',
                                            'connection : close\r\n',
                                            'Location : get.html\r\n',
                                            'Content-Type : text/html; charset=UTF-8\r\n',
                                            'Content-Length: ', clength, '\r\n')

        responseHeader_encode = responseHeader.encode(encoding = 'utf-8')
        
        clientsocket.send(responseStatus)
        clientsocket.send(responseHeader_encode)
        clientsocket.send(b'\r\n')
        clientsocket.send(responseData)

    if data_decode[5:18] == 'notfound.html':

        responseData = b'<html><body>404 Not Found</body></html>'
        clength = len(responseData)
        responseStatus = b'HTTP/1.0 404 Not Found\r\n'
        responseHeader = '%s%s%s%s%s%d%s' % ('Date : Mon, 07 Dec 2020 12:12:12 GMT\r\n',
                                            'Server : localhost\r\n',
                                            'connection : close\r\n',
                                            'Content-Type : text/html; charset=UTF-8\r\n',
                                            'Content-Length: ', clength, '\r\n')

        responseHeader_encode = responseHeader.encode(encoding = 'utf-8')

        clientsocket.send(responseStatus)
        clientsocket.send(responseHeader_encode)
        clientsocket.send(b'\r\n')
        clientsocket.send(responseData)

    if data_decode[5:14] == 'head.html':
        
        responseData = b''
        clength = len(responseData)
        responseStatus = b'HTTP/1.0 200 OK\r\n'
        responseHeader = '%s%s%s%s%d%s' % ('Date : Mon, 07 Dec 2020 12:12:12 GMT\r\n',
                                            'Server : localhost\r\n',
                                            'connection : close\r\n',
                                            'Content-Length:', clength, '\r\n')

        responseHeader_encode = responseHeader.encode(encoding = 'utf-8')

        clientsocket.send(responseStatus)
        clientsocket.send(responseHeader_encode)
        clientsocket.send(b'\r\n')
        clientsocket.send(responseData)

    if data_decode[5:14] == 'post.html':
        
        start = data_decode.find('Id : ')
        end = data_decode.find('\r', start)
        id_number = data_decode[start+5 : end-1]
        responseData ='<html><body>' + id_number + '</body></html>'
        responseData_encode = responseData.encode(encoding = 'utf-8')
        clength = len(responseData_encode)
        responseStatus = b'HTTP/1.0 200 OK\r\n'
        responseHeader = '%s%s%s%s%s%d%s' % ('Date : Mon, 07 Dec 2020 12:12:12 GMT\r\n',
                                            'Server : localhost\r\n',
                                            'connection : close\r\n',
                                            'Content-Type : text/html; charset=UTF-8\r\n',
                                            'Content-Length:', clength, '\r\n')

        responseHeader_encode = responseHeader.encode(encoding = 'utf-8')

        clientsocket.send(responseStatus)
        clientsocket.send(responseHeader_encode)
        clientsocket.send(b'\r\n')
        clientsocket.send(responseData_encode)
        
    clientsocket.close()
        
```

```python
while True:
    # accept a new client and get it's information
    (clientsocket, address) = s.accept()
    print("accept")
    # create a new thread to serve this new client
    # after the thread is created, it will start to execute 'target' function with arguments 'args' 
    threading.Thread(target = serveClient, args = (clientsocket, address)).start()
```
